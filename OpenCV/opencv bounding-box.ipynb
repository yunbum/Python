```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("./images/300.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray);
```

```python
img_blur = cv2.GaussianBlur(img_gray, (5, 5), 1)
plt.imshow(img_blur);
```

```python
ret, img_th = cv2.threshold(img_blur, 250, 255, cv2.THRESH_BINARY_INV)
plt.imshow(img_th);
```

```python
#image, contours, hierachy= cv2.findContours(img_th.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours, hierachy= cv2.findContours(img_th.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
rects = [cv2.boundingRect(each) for each in contours]
rects = [(x,y,w,h) for (x,y,w,h) in rects if ((w*h>200)and(w*h<500000))]

len(rects)
```

```python
for rect in rects:
    # Draw the rectangles
    cv2.rectangle(img, (rect[0], rect[1]), 
                  (rect[0] + rect[2], rect[1] + rect[3]), (0, 255, 0), 1)
```

```python
plt.figure(figsize=(15,12))
plt.imshow(img);
```
